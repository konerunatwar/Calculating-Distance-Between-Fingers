```python
#To Install open cv use * pip install opencv-python *
import cv2

#To Install mediapipe use * pip install mediapipe *
import mediapipe as mp

#To Install numpy use * pip install numpy *
import numpy as np

#Assigning Variables
mp_drawing = mp.solutions.drawing_utils
mp_hands = mp.solutions.hands

#Capturing video 
cap = cv2.VideoCapture(0) #Try using diffent number if  0 is giving youan error, zero is a webcam number

# * max_num_hands can be changed to 1 to dectect one hands *
with mp_hands.Hands(max_num_hands = 2,min_detection_confidence =0.7,min_tracking_confidence = 0.5) as hands:
    while cap.isOpened:
        ret,frame = cap.read()
        
        image = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
        image.flags.writeable = False
        
        results = hands.process(image)

        image.flags.writeable = True
        image = cv2.cvtColor(image,cv2.COLOR_RGB2BGR)
        
        if results.multi_hand_landmarks:
                for hand_landmarks in results.multi_hand_landmarks:
                    mp_drawing.draw_landmarks(image, hand_landmarks, mp_hands.HAND_CONNECTIONS,
                                             mp_drawing.DrawingSpec(color=(128,128,128), thickness=2, circle_radius=7), 
                                             mp_drawing.DrawingSpec(color=(192,192,192), thickness=2, circle_radius=7)
                                             )   
        try:
            whichhand = results.multi_handedness
            landmarks = hand_landmarks.landmark
            x = whichhand[0].classification[0].label
            j = round(whichhand[0].classification[0].score,5)
            
            
            # What i am doing here is taking the x,y coordinates of each finger tip 
            
            # WRIST
            b = [landmarks[mp_hands.HandLandmark.WRIST].x,landmarks[mp_hands.HandLandmark.WRIST].y]
            
            # PINKY_TIP
            b1 = [landmarks[mp_hands.HandLandmark.PINKY_MCP].x,landmarks[mp_hands.HandLandmark.PINKY_MCP].y]
            
            # INDEX_FINGER_TIP
            a1 = [landmarks[mp_hands.HandLandmark.INDEX_FINGER_TIP].x,landmarks[mp_hands.HandLandmark.INDEX_FINGER_TIP].y]
            
            # MIDDLE_FINGER_TIP
            a2 = [landmarks[mp_hands.HandLandmark.MIDDLE_FINGER_TIP].x,landmarks[mp_hands.HandLandmark.MIDDLE_FINGER_TIP].y]
            
            # RING_FINGER_TIP
            a3 = [landmarks[mp_hands.HandLandmark.RING_FINGER_TIP].x,landmarks[mp_hands.HandLandmark.RING_FINGER_TIP].y]
            
            # RING_FINGER_TIP
            a4 = [landmarks[mp_hands.HandLandmark.PINKY_TIP].x,landmarks[mp_hands.HandLandmark.PINKY_TIP].y]
            
            # RING_FINGER_TIP
            a5 = [landmarks[mp_hands.HandLandmark.THUMB_TIP].x,landmarks[mp_hands.HandLandmark.THUMB_TIP].y]
            
            # Now i am using distance between two points formula
            
            # DISTANCE BETWEEN INDEX_FINGER_TIP & WRIST
            distance1 = round((np.sqrt((a1[0]-b[0])**2 + (a1[1]-b[1])**2))*100,1)
            
            # printing distace on the screen 
            cv2.putText(image, str(distance1), tuple(np.multiply(a1, [640, 480]).astype(int)), cv2.FONT_HERSHEY_DUPLEX,0.5, (0, 255, 255),1, cv2.LINE_AA)
            
            # DISTANCE BETWEEN MIDDLE_FINGER_TIP & WRIST
            distance2 = round((np.sqrt((a2[0]-b[0])**2 + (a2[1]-b[1])**2))*100,1)
            
            # printing distace on the screen 
            cv2.putText(image, str(distance2), tuple(np.multiply(a2, [640, 480]).astype(int)), cv2.FONT_HERSHEY_DUPLEX,0.5, (145, 0, 253),1, cv2.LINE_AA)
            
            # DISTANCE BETWEEN RING_FINGER_TIP & WRIST
            distance3 = round((np.sqrt((a3[0]-b[0])**2 + (a3[1]-b[1])**2))*100,1)
            
            # printing distace on the screen 
            cv2.putText(image, str(distance3), tuple(np.multiply(a3, [640, 480]).astype(int)), cv2.FONT_HERSHEY_DUPLEX,0.5, (255, 255, 0),1, cv2.LINE_AA)
        
            # DISTANCE BETWEEN PINKY_TIP & WRIST
            distance4 = round((np.sqrt((a4[0]-b[0])**2 + (a4[1]-b[1])**2))*100,1)
            
            # printing distace on the screen 
            cv2.putText(image, str(distance4), tuple(np.multiply(a4, [640, 480]).astype(int)), cv2.FONT_HERSHEY_DUPLEX,0.5, (255, 100, 255),1, cv2.LINE_AA)
            
            # DISTANCE BETWEEN THUMB_TIP & PINKY_MCP
            distance5 = round((np.sqrt((a5[0]-b1[0])**2 + (a5[1]-b1[1])**2))*100,1)
            
            # printing distace on the screen 
            cv2.putText(image, str(distance5), tuple(np.multiply(a5, [640, 480]).astype(int)), cv2.FONT_HERSHEY_DUPLEX,0.5, (255, 255, 120),1, cv2.LINE_AA)
            
            cv2.putText(image,x,tuple(np.multiply(b, [660, 490]).astype(int)),cv2.FONT_HERSHEY_DUPLEX, 0.5, (255,255,255), 1, cv2.LINE_AA)
            cv2.putText(image,str(j),tuple(np.multiply(b, [660, 510]).astype(int)),cv2.FONT_HERSHEY_DUPLEX, 0.5, (255,255,255), 1, cv2.LINE_AA)
            
            # MADE MY OWN SIGNS USING THE DISTANCES
            
            # STEP 1 : SPREAD ALL YOUR FINGERS, ex show 5 and note all the distaces. This will be the maximum distance from the WRIST.
            # STEP 2 : NOW USE YOUR SIGN and note the distance.

            # SOME EXAMPLES 
            
            # âœŠ / Zero
            if (distance1 < 40) and (distance2 < 40) and (distance3 < 40) and (distance4 < 40) and (distance5 < 30):
                sine = '0'
            
            # â˜ / One
            if (distance1 > 40) and (distance2 < 40) and (distance3 < 40) and (distance4 < 40) and (distance5 < 30):
                sine = '1'
                
            # âœŒï¸ / Two
            if (distance1 > 40) and (distance2 > 40) and (distance3 < 40) and (distance4 < 40) and (distance5 < 30):
                sine = '2'
            
            # Three
            if (distance1 > 40) and (distance2 > 40) and (distance3 > 40) and (distance4 < 40) and (distance5 < 30):
                sine = '3'
            
            # Four
            if (distance1 > 40) and (distance2 > 40) and (distance3 > 40) and (distance4 > 40) and (distance5 < 30):
                sine = '4'
            
            # ðŸ– / Five
            if (distance1 > 40) and (distance2 > 40) and (distance3 > 40) and (distance4 > 40) and (distance5 > 30):
                sine = '5'
                
            # ðŸ¤˜ / Rock On
            if (distance1 > 40) and (distance2 < 40) and (distance3 < 40) and (distance4 > 40) and (distance5 > 30):
                sine = 'RockOn'
                
            
            
            else:
                pass

            
            
            
            cv2.putText(image,sine, 
                    (15,35), 
                    cv2.FONT_HERSHEY_DUPLEX, 0.8, (255,255,255), 2, cv2.LINE_AA)
            

        except:
            pass
        
    
        cv2.imshow('', image)

        if cv2.waitKey(10) & 0xFF == ord('q'):
            break
            
    cap.release()
    cv2.destroyAllWindows()
```
